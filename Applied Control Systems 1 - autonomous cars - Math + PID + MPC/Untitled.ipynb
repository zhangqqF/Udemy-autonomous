# matplotlib动画

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np


fig = plt.figure()


x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)
ax = fig.add_subplot(1, 1, 1)
sin, = ax.plot([], [], color='cornflowerblue', lw=3)


# 静图
picture, = ax.plot(x, y, lw=3)
plt.show()
```

```python
def anim_upt(frame):
    print(frame, '\n')
    # return
    sin.set_data(x[:frame-int(frame/len(x))], y[:frame-int(frame/len(x))])
#     print(x[:frame], y[:frame])
    return sin,

anim = animation.FuncAnimation(fig, anim_upt, frames=len(x), interval=20, blit=True)
plt.show()
```
